```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

## Simple Graphs

```python
#list storing date in string format
date=["25/12","26/12","27/12"]

#list storing temperature values
temp=[8.5,10.5,6.8]

#create a figure plotting temp versus date         (X,y)
plt.plot(date, temp)

#show the figure
plt.show()
```

```python
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]
plt.plot(date, temp, marker='D')
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Date VS Temperature")
plt.grid(True)
plt.yticks(temp)
plt.show()
```

## Customising Plots

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
df=pd.DataFrame({"height":height,"weight":weight})
plt.xlabel("Weight(in kg)")
plt.ylabel("Height(in cm)")
plt.title("Avg Height over Avg Weight")
plt.plot(df.weight, df.height, marker='o', markersize=10, color='green', linewidth=2, linestyle='dashdot')
```

## Scatter Plots

```python
discount= np.array([10,20,30,40,50])
saleInRs=np.array([40000,45000,48000,50000,100000])
plt.scatter(x=discount,y=saleInRs)
plt.title('Sales Vs Discount')
plt.xlabel('Discount offered')
plt.ylabel('Sales in Rs')
plt.show()
```

```python
size=discount*10 # => The size of the bubble can also be used to reflect a value.
plt.scatter(x=discount,y=saleInRs,s=size,color='red',linewidth=3,marker='*',edgecolor='blue')
plt.title('Sales Vs Discount')
plt.xlabel('Discount offered')
plt.ylabel('Sales in Rs')
plt.show()
```

## Box Plots

```python
df = pd.DataFrame({
    'Sunny Bunny Resort': [4.75, 2.5,3.5,4,1.5],
    'Happy Lucky Resort': [3,4,2.5,2,4.5],
    'Breezy Windy Resort': [4.5,2,3,3.5,1]
}, index=[i for i in range(2014,2019)])
df.index.name = 'Year'
df
```

```python
df.plot(kind='box',title='Compare Resorts', xlabel='Resorts', ylabel='Rating (5 years)')
plt.show()
```

```python
df.plot(kind='box',title='Compare Resorts', xlabel='Resorts', ylabel='Rating (5 years)', color='red', vert=False)
plt.show()
```

## Pie Charts

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97], 'radius': [2439.7, 6051.8, 6378.1]}, index=['Mercury', 'Venus', 'Earth'])
df.plot(kind='pie',y='mass')
plt.show()
```

```python
df.plot(
    kind='pie',
    y='mass',
    legend=False,
    colors=['gray','y','g'],
    explode=[0.1,0.1,0.1],
    autopct="%.2f"
)
plt.show()
```
