# Simplr Imputer 

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
df_num = df_train[num_attribs]
imputer.fit(df_num)
df_num_array = imputer.transform(df_num)
df_train[num_attribs] = pd.DataFrame(df_num_array, columns=df_num.columns,index=df_num.index)
df_train.head(10)
```

```python
df_num = df_test[num_attribs]
imputer.fit(df_num)
df_num_array = imputer.transform(df_num)
df_test[num_attribs] = pd.DataFrame(df_num_array, columns=df_num.columns,index=df_num.index)
df_test.head(10)
```

```python
df=pd.concat([df_train,df_test],axis=0)
```

```python
df.fillna(df.select_dtypes(include='object').mode().iloc[0], inplace=True)
```

```python
df.info()
```

# Encoder

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
df_cat = df[cat_attribs]
df_cat_encoded = ordinal_encoder.fit_transform(df_cat)
df[cat_attribs] = pd.DataFrame(df_cat_encoded, columns=df_cat.columns,index=df_cat.index)
```

```python
df_Train=df.iloc[:1422,:]
df_Test=df.iloc[1422:,:] ## to predict and send on ##
```

```python
df.isnull().sum()
```

```python
X_Train = df_Train.drop("SalePrice",axis=1)
y_Train = df_Train["SalePrice"]
X_Test = df_Test.drop("SalePrice",axis=1)
# y_Test = df_test["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_cv, y_train, y_cv = train_test_split(X_Train, y_Train, test_size=1/3, random_state=100000)
```

# MODEL SELECTION (Regression)

```python
from sklearn.metrics import r2_score

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import ExtraTreesRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.ensemble import VotingRegressor
```

```python
def model_selection(X_train, X_cv, y_train, y_cv):
    
    rnd_reg = RandomForestRegressor(random_state = 42)
    rnd_reg.fit(X_train, y_train)
    y_pred_train = rnd_reg.predict(X_train)
    rnd_score_train = r2_score(y_train, y_pred_train) ##
    y_pred_cv = rnd_reg.predict(X_cv)
    rnd_score_cv = r2_score(y_cv, y_pred_cv) ##
    
#     sgd_reg_1 = SGDRegressor(penalty='l2')
#     sgd_reg_1.fit(X_train, y_train)
#     y_pred_train = sgd_reg_1.predict(X_train)
#     sgd_reg_train_score1 = r2_score(y_train, y_pred_train) ##
#     y_pred_cv = sgd_reg_1.predict(X_cv)
#     sgd_reg_score_cv1 = r2_score(y_cv, y_pred_cv) ##
    
    ext_rnd_reg = ExtraTreesRegressor(random_state=42)
    ext_rnd_reg.fit(X_train, y_train)
    y_pred_train = ext_rnd_reg.predict(X_train)
    ext_rnd_reg_score_train = r2_score(y_train, y_pred_train) ##
    y_pred_cv = ext_rnd_reg.predict(X_cv)
    ext_rnd_reg_score_cv = r2_score(y_cv, y_pred_cv) ##
    
    xgb_reg = XGBRegressor(random_state = 42)
    xgb_reg.fit(X_train, y_train)
    y_pred_train = xgb_reg.predict(X_train)
    xgb_reg_train_score = r2_score(y_train, y_pred_train) ##
    y_pred_cv = xgb_reg.predict(X_cv)
    xgb_reg_score_cv = r2_score(y_cv, y_pred_cv) ##
    
    gbrt = GradientBoostingRegressor(learning_rate=0.1,random_state = 42)
    gbrt.fit(X_train, y_train)
    y_pred_train = gbrt.predict(X_train)
    gbrt_train_score = r2_score(y_train, y_pred_train) ##
    y_pred_cv = gbrt.predict(X_cv)
    gbrt_score_cv = r2_score(y_cv, y_pred_cv) ##
    
#     sgd_reg_2 = SGDRegressor(penalty='l1')
#     sgd_reg_2.fit(X_train, y_train)
#     y_pred_train = sgd_reg_2.predict(X_train)
#     sgd_reg_train_score2 = r2_score(y_train, y_pred_train) ##
#     y_pred_cv = sgd_reg_2.predict(X_cv)
#     sgd_reg_score_cv2 = r2_score(y_cv, y_pred_cv) ##
    
#     sgd_reg_3 = SGDRegressor(penalty='elasticnet')
#     sgd_reg_3.fit(X_train, y_train)
#     y_pred_train = sgd_reg_3.predict(X_train)
#     sgd_reg_train_score3 = r2_score(y_train, y_pred_train) ##
#     y_pred_cv = sgd_reg_3.predict(X_cv)
#     sgd_reg_score_cv3 = r2_score(y_cv, y_pred_cv) ##
    
    svm_reg = SVR(kernel='linear')
    svm_reg.fit(X_train, y_train)
    y_pred_train = svm_reg.predict(X_train)
    svm_reg_score_train = r2_score(y_train, y_pred_train) ##
    y_pred_cv = svm_reg.predict(X_cv)
    svm_reg_score_cv = r2_score(y_cv, y_pred_cv) ##
    
    estimators = [('rf',rnd_reg),('ext',ext_rnd_reg),('xgb',xgb_reg),('gbrt',gbrt),('svm',svm_reg)]
    voting_reg = VotingRegressor(estimators=estimators)
    voting_reg.fit(X_train, y_train)
    y_pred_train = voting_reg.predict(X_train)
    voting_reg_train_score = r2_score(y_train, y_pred_train) ##
    y_pred_cv = voting_reg.predict(X_cv)
    voting_reg_score_cv = r2_score(y_cv, y_pred_cv) ##
    
    
    
    Models = ['RandomForestRegressor','ExtraTreesRegressor','XGBRegressor','GradientBoostingRegressor','SVR', 'VotingRegressor']
    train_scores = [rnd_score_train,ext_rnd_reg_score_train,xgb_reg_train_score,gbrt_train_score,svm_reg_score_train,voting_reg_train_score]
    cv_scores = [rnd_score_cv,ext_rnd_reg_score_cv,xgb_reg_score_cv,gbrt_score_cv,svm_reg_score_cv,voting_reg_score_cv]
    
    dic = {'Models': Models, 'train_scores':train_scores, 'cv_scores':cv_scores}
    table = pd.DataFrame(dic)
    print(table)
```

```python
model_selection(X_train, X_cv, y_train, y_cv)
```
